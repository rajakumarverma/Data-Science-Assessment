```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
# Need to keep name column first letter as Capital
df['name'] = df['name'].str.title()
```

```python
df.head()
```

```python
# Need to merge name and surname column 
df['full_name'] = df['name'] + ' ' + df['surname']
```

```python
df.head()
```

```python
# To keep separate alternate number in new column
df['Alternate Number'] = df['mobile'].apply(lambda x: x.split(',')[1] if ',' in x else None)
```

```python
df.head()
```

```python
# There are two contact number in mobile column, need to keep separate
df['mobile'] = df['mobile'].apply(lambda x: x.split(',')[0])
```
